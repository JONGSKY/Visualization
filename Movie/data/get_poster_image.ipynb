```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse
import pandas as pd
import time
```

```python
def get_poster_img(movieNm):
    # get the image source
    img = driver.find_element_by_xpath('//*[@id="content"]/div[1]/div[2]/div[2]/a/img')
    src = img.get_attribute('src')

    # download the image
    urllib.request.urlretrieve(src, "poster/"+movieNm+".jpg")
```

```python
movie_df = pd.read_csv('movie_data.csv')
```

```python
movie_df.head()
```

```python
driver = webdriver.Chrome('./chromedriver')
basic_url = 'https://movie.naver.com/'
driver.get(basic_url)
time.sleep(2)

# alert 창의 '확인'을 클릭합니다.
alert = driver.switch_to.alert
alert.accept()
time.sleep(2)

data_list = []
for movieNm in movie_df['movieNm']:
    data_dict = {}
    data_dict['movieNm'] = [movieNm]
    
    try:
        url = 'https://movie.naver.com/movie/search/result.nhn?query='+urllib.parse.quote(movieNm)+'&section=all&ie=utf8'
        driver.get(url)
        time.sleep(2)
        driver.find_element_by_xpath('//*[@id="old_content"]/ul[2]/li/p/a/img').click()
        data_dict['movieUrl'] = [driver.current_url]
        time.sleep(2)
        get_poster_img(movieNm)
    except:
        data_dict['movieUrl'] = ["수작업 필요"]
        print('영화제목 : '+movieNm+' 은 저장하지못했습니다. 확인해주세요.')
        
    data_df = pd.DataFrame(data_dict)
    data_list.append(data_df)
driver.close()
```

```python
all_df = pd.concat(data_list).reset_index(drop=True)
all_df.head()
```
