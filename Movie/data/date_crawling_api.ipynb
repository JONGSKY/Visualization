```python
import requests
import json
import pandas as pd
import datetime
from dateutil.relativedelta import *
```

```python
with open('key_info.json') as json_file:
    json_data = json.load(json_file)
    key = json_data['key']
```

```python
numdays = 120
```

```python
base = datetime.datetime(2020, 9, 1)
date_list = [(base + relativedelta(months=-x)).strftime('%Y%m%d') for x in range(numdays)]
```

```python
data_list = []
for date in date_list:
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key='+key+'&targetDt='+date
    data = json.loads(requests.get(url).text)['boxOfficeResult']['dailyBoxOfficeList']
    movie_data = pd.DataFrame(data)
    movie_data['searchdate'] = date
    data_list.append(movie_data)
```

```python
all_df = pd.concat(data_list).reset_index(drop=True)
all_df.head()
```

```python
all_df.to_csv('movie_data.csv', index=False)
```
